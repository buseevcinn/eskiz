## VERİ ANALİZİ UYGULAMASI : Büyük bir araç veri seti üzerinden temel python analizi

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression #Scikit-learn kütüphanesinden LinearRegression sınıfını içe aktarır.
from sklearn.model_selection import train_test_split #Veriyi eğitim (train) ve test olmak üzere ikiye ayırmak için kullanılır.
from sklearn.metrics import mean_squared_error, r2_score #Modelin doğruluğunu ölçmek için kullanılan değerlendirme metriklerini içe aktarır 
```

```python
data = pd.read_csv("/Users/buse/Desktop/ford.csv")
data.head()
```

### Dağılım Grafiği : Fiyat(price) ve Motor Hacmi(engineSize)

```python
sns.scatterplot(data=data, x="engineSize", y="price")
plt.title("Motor Hacmi ve Fiyat İlişkisi")
plt.xlabel("Motor Hacmi")
plt.ylabel("Fiyat")
plt.show
```

### Model Başına Ortalama Fiyat (bar grafiği)

```python
plt.figure(figsize=(12,6))
sns.barplot(data=data, x="model", y="price", estimator="mean", errorbar=None)
plt.title("Modellere Göre Ortalama Fiyat")
plt.ylabel("Ortalama Fiyat")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

### Yıllara Göre Ortalama Fiyat Grafiği

```python
plt.figure(figsize=(10,5))
sns.lineplot(data=data, x="year", y="price", estimator="mean", errorbar=None)
plt.title("Yıllara Göre Ortalama Fiyat")
plt.ylabel("Ortalama Fiyat")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

### Yakıt Türüne Göre Fiyat Dağılımı 

```python
sns.boxplot(data=data, x="fuelType", y="price")
plt.title("Yakıt Türüne Göre Fiyat Dağılımı")
plt.xticks(rotation=45)
plt.show()
```

## REGRESYON ANALİZİ (FİYAT TAHMİNİ)

```python
#Özellikler ve değişken tanıma
x=data[["year", "mileage", "mpg", "engineSize", "tax"]]
y=data[["price"]]

#Eğitim ve test verisini ayırma
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0)

#Model oluşturma ve eğitme
model= LinearRegression()
model.fit(x_train, y_train)

#Tahmin Yapma
tahmin= model.predict(x_test)

#Performans Değerlendirme
r2= r2_score(y_test, tahmin)
mse= mean_squared_error(y_test, tahmin)
rmse = mse ** 0.5

#Modelin başarısını değerlendirme
print("R kare Skoru: ", r2)
print("MSE: ", mse)
print("Karekök Ortalama Hata(RMSE): ",rmse)
```
